```python
from pathlib import Path
import sys

NOTEBOOK_DIR = Path.cwd()
if (NOTEBOOK_DIR / 'train_prefilter.py').exists():
    sys.path.insert(0, str(NOTEBOOK_DIR))
else:
    baseline_dir = Path('GRACE-improve/baseline').resolve()
    sys.path.insert(0, str(baseline_dir))
```

```python
from train_prefilter import train_prefilter
```

```python
DATASET_NAME = 'devign'
BATCH_SIZE = 128
MAX_TOKENS = 40000
SEQUENCE_LENGTH = 512
EMBEDDING_DIM = 128
FILTERS = 128
DENSE_UNITS = 128
DROPOUT_RATE = 0.2
EPOCHS = 12
LEARNING_RATE = 1e-3
RANDOM_SEED = 42

LOG_PROGRESS = True
KERAS_VERBOSE = 0
BATCH_LOG_INTERVAL = None  # None = auto, 0 = chi log theo epoch, N = log moi N batch
```

```python
print(
    f'Train prefilter on {DATASET_NAME} | epochs={EPOCHS} | batch_size={BATCH_SIZE} | '
    f'keras_verbose={KERAS_VERBOSE} | batch_log_interval={BATCH_LOG_INTERVAL}'
)
```

```python
summary = train_prefilter(
    dataset_name=DATASET_NAME,
    batch_size=BATCH_SIZE,
    max_tokens=MAX_TOKENS,
    sequence_length=SEQUENCE_LENGTH,
    embedding_dim=EMBEDDING_DIM,
    filters=FILTERS,
    dense_units=DENSE_UNITS,
    dropout_rate=DROPOUT_RATE,
    epochs=EPOCHS,
    learning_rate=LEARNING_RATE,
    random_seed=RANDOM_SEED,
    verbose=KERAS_VERBOSE,
    log_progress=LOG_PROGRESS,
    batch_log_interval=BATCH_LOG_INTERVAL,
)
summary
```
