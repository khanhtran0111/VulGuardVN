# Loc va xuat cac dong vul == 1
Notebook nay doc file CSV, loc cac dong co cot `vul` bang 1, va ghi ra file CSV moi.

```python
from pathlib import Path
import pandas as pd

# Duong dan mac dinh theo cau truc workspace: GRACE-main -> ../dataset
input_csv = Path('..') / 'dataset' / 'MSR_data_cleaned.csv'
output_csv = Path('.') / 'MSR_data_cleaned_vul_1.csv'

df = pd.read_csv(input_csv)

if 'vul' not in df.columns:
    raise KeyError("Khong tim thay cot 'vul' trong file CSV.")

df_vul_1 = df[df['vul'] == 1].copy()
df_vul_1.to_csv(output_csv, index=False, encoding='utf-8-sig')

print(f'Tong so dong ban dau: {len(df)}')
print(f'So dong co vul == 1: {len(df_vul_1)}')
print(f'Da xuat file: {output_csv.resolve()}')
```
